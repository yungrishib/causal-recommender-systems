```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
exposure_df = pd.read_csv("../data/exposure_over_time.csv")
exposure_df.head()
```

```python
print(exposure_df.columns)
print(exposure_df.shape)
```

```python
plt.figure()
plt.plot(exposure_df["round"], exposure_df["A"], label="Group A")
plt.plot(exposure_df["round"], exposure_df["B"], label="Group B")
plt.xlabel("Recommendation Round")
plt.ylabel("Top-K Exposure Count")
plt.title("Exposure Inequality Over Time")
plt.legend()
plt.show()
```

```python
exposure_df["bias_gap"] = exposure_df["A"] - exposure_df["B"]
exposure_df[["round", "bias_gap"]]
```
